## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup


#browser = webdriver.Chrome(executable_path='chromedriver')
browser = webdriver.Chrome('../chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 每個兩秒鐘自動往下滑
def scroller(browser):
    import time
    for i in range(150):
        time.sleep(2)
        browser.execute_script("window.scrollTo(0, 10000);")
```

```python
scroller(browser)
```

```python
# 取得資料，丟到 BeautifulSoup 解析
from datetime import datetime, timedelta

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
dateFormat='%Y/%m/%d %H:%M'
temp=str(datetime.now())
y=temp[0:4]
m=temp[5:7]
d=temp[8:10]

datetime_start = datetime.strptime(y+'/'+m+'/'+d+' 00:00', dateFormat)
datetime_end = datetime.strptime(y+'/'+m+'/'+str(int(d)+1)+' 00:00', dateFormat)
#print(datetime_start,datetime_end)

for d in soup.find(class_="part_list_2").find_all('h3'):
    cellTime = datetime.strptime(d.find(class_="date").text,dateFormat)
    if(cellTime>=datetime_start and cellTime<datetime_end):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta

start=datetime.now()-timedelta(hours=2)

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
dateFormat='%Y/%m/%d %H:%M'

datetime_obj = datetime.now()-timedelta(hours=2)
for d in soup.find(class_="part_list_2").find_all('h3'):
    if(datetime.strptime(d.find(class_="date").text,dateFormat)>=datetime_obj):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
