```python
import re
def regMatch(regex, input_text):
    #print(table.find(class_='no-alt'))
    pattern = re.compile(regex)

    # 使轉換後的pattern，來測試是否匹配
    result = re.search(pattern, input_text)
    
    if result :
        return True;
    else :
        return False;
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')

table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
data=dict()
title=list()

for cellData in table.find_all("td", attrs={"style":"white-space:nowrap;"}):
    title.append(cellData.text)

count=0
for d in table.find_all("td"):
    if len(d.find_all(text=title[count]))>0 :
        if count<(len(title)-1):
            count+=1
        data[title[count]]=dict()
    if regMatch('(^\d{4}/\d{2}$)',d.text):
        data[title[count]].update({d.text:d.next_sibling.text})   
    
    
data
```
