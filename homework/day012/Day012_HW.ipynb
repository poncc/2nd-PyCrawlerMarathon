### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find('',class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

dataDict=dict();
data = {}
news={}
dateList={}

for d in soup.find('',class_="part_list_2").find_all('h3'):
    tag=d.find(class_="tag").text;
    newsTitle=d.find_all('a')[-1].text;
    if tag not in dataDict.keys():
        dataDict[tag]=list();
        dataDict[tag].append(newsTitle)
    else:
        dataDict[tag].append(newsTitle)
print(dataDict)
```

### ③ 哪一個類別的文章最多

```python
maxCount=0
maxKey=''
for key,value in dataDict.items():
    if(len(value)>maxCount):
        maxCount=len(value)
        maxKey=key
        
print(maxKey,maxCount)
    
```
